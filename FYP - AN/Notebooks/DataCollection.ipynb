#### Data Collection

```python
import os
import pandas as pd
import numpy as np
```

#### Basic

```python
path = '../Data/Basic/'
basicFiles = os.listdir(path)
for i in range(len(basicFiles)):
    basicFiles[i] = path + basicFiles[i]
    
basicFiles
```

```python
basic = pd.DataFrame()

for fileName in basicFiles:
    data = pd.read_excel(fileName, sheet_name = 'Final')
    basic = pd.concat([basic,data], axis=0)

#basic = basic.set_index(['type']) 
basic
```

```python
basic.to_excel('../Data/FinalData/BasicFinal.xlsx')
```
